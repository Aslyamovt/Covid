```python
#Функция рассчета числа выздровевших за n-й день при гамма распределении
from scipy.special import gamma
def calculate_recs_gamma(n,a_inf,b_inf,k_inf,teta_inf,a_dead,b_dead,k_dead,teta_dead,a_inc,b_inc,k_inc,teta_inc):
    recs_pred = 0
    for j in range(n+1):
        inf_ch = 0
        if j>0:
            inf_ch = (max(0,(b_inf*((j-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real) - 
            max(0,(b_inf*((j-1-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-1-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real))
        else:
            inf_ch = max(0,(b_inf*((j-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real)
        recs_pred+= max(0,(max(0,(b_inc*((j-a_inc)**(k_inc-1)*
                              ((math.e**(-(j-a_inc)/teta_inc))/(teta_inc**k_inc*gamma(k_inc))))).real) - 
                     inf_ch - max(0,(b_dead*((j-a_dead)**(k_dead-1)*
                              ((math.e**(-(j-a_dead)/teta_dead))/(teta_dead**k_dead*
                                                                  gamma(k_dead))))).real)))
    return recs_pred
```

```python
#Функция рассчета числа выздровевших за n-й день при гамма распределении для числа болеющих и
#прироста числа заболевших, и нормальном распределении при приросте числа умерших
#Название длинное, поэтому дальше буду называеть его иногда первым смешанным случаем
from scipy.special import gamma
def calculate_recs(n,a_inf,b_inf,k_inf,teta_inf,a_dead,b_dead,c_dead,a_inc,b_inc,k_inc,teta_inc):
    recs_pred = 0
    for j in range(n+1):
        inf_ch = 0
        if j>0:
            inf_ch = (max(0,(b_inf*((j-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real) - 
            max(0,(b_inf*((j-1-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-1-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real))
        else:
            inf_ch = max(0,(b_inf*((j-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real)
        recs_pred+= max(0,(max(0,(b_inc*((j-a_inc)**(k_inc-1)*
                              ((math.e**(-(j-a_inc)/teta_inc))/(teta_inc**k_inc*gamma(k_inc))))).real) - 
                     inf_ch - b_dead*(math.e**(-c_dead*(j-a_dead)**2))))
    return recs_pred
```

```python
#Функция рассчета числа выздровевших за n-й день при гамма распределении для числа болеющих 
# и нормальном распределении при приросте числа умерших и приросте числа заболевших
#Название длинное, поэтому дальше буду называеть его иногда вторым смешанным случаем
from scipy.special import gamma
def calculate_recs2(n,a_inf,b_inf,k_inf,teta_inf,a_dead,b_dead,c_dead,a_inc,b_inc,c_inc):
    recs_pred = 0
    for j in range(n+1):
        inf_ch = 0
        if j>0:
            inf_ch = (max(0,(b_inf*((j-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real) - 
            max(0,(b_inf*((j-1-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-1-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real))
        else:
            inf_ch = max(0,(b_inf*((j-a_inf)**(k_inf-1)*
                              ((math.e**(-(j-a_inf)/teta_inf))/(teta_inf**k_inf*gamma(k_inf))))).real)
        recs_pred+= max(0,max(0,(b_inc*(math.e**(-c_inc*(j-a_inc)**2)) - 
                     inf_ch - b_dead*(math.e**(-c_dead*(j-a_dead)**2)))))
    return recs_pred
```
